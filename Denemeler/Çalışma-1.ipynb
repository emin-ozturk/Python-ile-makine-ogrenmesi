```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
data = pd.read_csv('ds_salaries.csv')
data.info()
data.head(3)
```

```python
y = data['salary']
x = data.drop('salary', axis=1)
X = x.values
Y = y.values
```

```python
print('Sütun bilgileri')
print("experience_level {0}".format(data['experience_level'].unique()))
print("employment_type {0}".format(data['employment_type'].unique()))
print("job_title {0}".format(data['job_title'].unique()))
print("salary_currency {0}".format(data['salary_currency'].unique()))
print("employee_residence {0}".format(data['employee_residence'].unique()))
print("company_location {0}".format(data['company_location'].unique()))
print("company_size {0}".format(data['company_size'].unique()))
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
ohe = OneHotEncoder()

work_year = x[['work_year']]
salary_in_usd = x[['salary_in_usd']]
remote_ratio = x[['remote_ratio']]
temp_experience_level = x[['experience_level']].values
temp_employment_type = x[['employment_type']].values
temp_job_title = x[['job_title']].values
temp_salary_currency = x[['salary_currency']].values
temp_employee_residence = x[['employee_residence']].values
temp_company_location = x[['company_location']].values
temp_company_size = x[['company_size']].values

temp_experience_level[:,0] = le.fit_transform(temp_experience_level)
temp_experience_level = ohe.fit_transform(temp_experience_level).toarray()

temp_employment_type[:,0] = le.fit_transform(temp_employment_type)
temp_employment_type = ohe.fit_transform(temp_employment_type).toarray()

temp_job_title[:,0] = le.fit_transform(temp_job_title)
temp_job_title = ohe.fit_transform(temp_job_title).toarray()

temp_salary_currency[:,0] = le.fit_transform(temp_salary_currency)
temp_salary_currency = ohe.fit_transform(temp_salary_currency).toarray()

temp_employee_residence[:,0] = le.fit_transform(temp_employee_residence)
temp_employee_residence = ohe.fit_transform(temp_employee_residence).toarray()

temp_company_location[:,0] = le.fit_transform(temp_company_location)
temp_company_location = ohe.fit_transform(temp_company_location).toarray()

temp_company_size[:,0] = le.fit_transform(temp_company_size)
temp_company_size = ohe.fit_transform(temp_company_size).toarray()

temp_company_size
```

```python
experience_level = pd.DataFrame(data=temp_experience_level, columns=['EN','EX','MI','SE'])
employment_type = pd.DataFrame(data=temp_employment_type)
job_title = pd.DataFrame(data=temp_job_title)
salary_currency = pd.DataFrame(data=temp_salary_currency)
employee_residence = pd.DataFrame(data=temp_employee_residence)
company_location = pd.DataFrame(data=temp_company_location)
temp_company_size = pd.DataFrame(data=temp_company_size)

new_data = pd.concat([
    work_year,
    experience_level,
    employment_type,
    job_title,
    salary_currency,
    employee_residence,
    company_location,
    temp_company_size,
    salary_in_usd,
    remote_ratio],
    axis=1)
new_data.info()
new_data.columns = new_data.columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(new_data, y, test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train, y_train)
y_pred = dtr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
new_data.corr()

import statsmodels.api as sm
model=sm.OLS(y_pred,x_test)
print(model.fit().summary())
```
