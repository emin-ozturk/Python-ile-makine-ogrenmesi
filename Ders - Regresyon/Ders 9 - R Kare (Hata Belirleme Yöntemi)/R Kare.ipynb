```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
```

```python
veriler = pd.read_csv("maaslar.csv")
veriler
```

```python
x = veriler.iloc[:,1:2]
y = veriler.iloc[:,2:]

X = x.values
Y = y.values
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X,Y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y, color="red")
plt.plot(x,lr.predict(X), color="blue")
plt.show()
```

```python
print(r2_score(Y, lr.predict(X)))
```

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree = 3)

#verileri polinomal hale çeviriyoruz
x_poly = pf.fit_transform(X)
x_poly
#x^0, x^1,x^2
#X değeri 1-10 arası olsuğu için karesi de 1-100 arası
```

```python
lr2 = LinearRegression()
lr2.fit(x_poly, y)

plt.scatter(X,Y)
plt.plot(X,lr2.predict(pf.fit_transform(X)), color="blue")
plt.show()
```

```python
print(lr.predict([[11]]))
print(lr2.predict(pf.fit_transform([[11]])))
```

```python
print(r2_score(Y, lr2.predict(pf.fit_transform(X))))
```

# SVR

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc1 = StandardScaler()
sc2 = StandardScaler()

x_olcekli = sc1.fit_transform(X)
y_olcekli = sc2.fit_transform(Y)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel = "rbf")
svr.fit(x_olcekli, y_olcekli)

plt.scatter(x_olcekli, y_olcekli)
plt.plot(x_olcekli, svr.predict(x_olcekli))
```

```python
print(svr.predict([[11]]))
print(svr.predict([[6.6]]))
```

```python
print(r2_score(y_olcekli, svr.predict(x_olcekli)))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(random_state=0)
dt.fit(X,Y)

plt.scatter(X,Y, color="red")
plt.plot(X, dt.predict(X), color="blue")
plt.show()
```

```python
print(dt.predict([[11]]))
print(dt.predict([[6.6]]))
```

```python
print(r2_score(Y, dt.predict(X)))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#n_estimators kaç farklı ağaç olacak
rf = RandomForestRegressor(n_estimators = 10,  random_state=0)
rf.fit(X,Y.ravel())
print(rf.predict([[6.6]]))
```

```python
plt.scatter(X,Y, color="red")
plt.plot(X, rf.predict(X), color="blue")
```

# R Kare

```python
print(r2_score(Y, rf.predict(X)))
```
