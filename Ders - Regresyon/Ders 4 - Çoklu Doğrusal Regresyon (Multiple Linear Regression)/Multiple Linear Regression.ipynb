```python
import numpy as np
import pandas as pd
```

```python
veriler = pd.read_csv("veriler.csv")
veriler
```

```python
ulke = veriler.iloc[:,0:1].values
cinsiyet = veriler.iloc[:,-1:].values
print(ulke, cinsiyet)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

ulke[:,0] = le.fit_transform(ulke)
cinsiyet[:,-1] = le.fit_transform(cinsiyet)

print(ulke, cinsiyet)
```

```python
ohe = preprocessing.OneHotEncoder()

ulke = ohe.fit_transform(ulke).toarray()
cinsiyet = ohe.fit_transform(cinsiyet).toarray()

print(ulke, cinsiyet)
```

```python
sonuc = pd.DataFrame(data=ulke, index=range(22), columns=["fr", "tr", "us"])
sonuc2 = pd.DataFrame(data=veriler.iloc[:,1:4], index=range(22), columns=["boy", "kilo", "yas"])
sonuc3 = pd.DataFrame(data=cinsiyet[:,:1], index=range(22), columns=["cinsiyet"])
```

```python
s = pd.concat([sonuc, sonuc2], axis=1)
s2 = pd.concat([s, sonuc3], axis=1)
s2
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(s, cinsiyet[:,:1], test_size=0.33, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train, y_train)

y_pred = regression.predict(x_test)
y_pred
```

```python
boy = s2.iloc[:,3:4].values
boy
```

```python
sol = s2.iloc[:,:3]
sag = s2.iloc[:,4:]
veri = pd.concat([sol, sag], axis=1)
veri
```

```python
x_train, x_test, y_train, y_test = train_test_split(veri, boy, test_size=0.33, random_state=0)
```

```python
regression = LinearRegression()
regression.fit(x_train, y_train)

y_pred = regression.predict(x_test)
y_pred
```
