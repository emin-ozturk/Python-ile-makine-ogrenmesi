```python
import numpy as np
import pandas as pd
```

```python
veriler = pd.read_csv("odev_tenis.csv")
veriler
```

# Veri Hazırlama

```python
outlook = veriler.iloc[:, 0:1].values
windy = veriler.iloc[:, 3:4].values
play = veriler.iloc[:, 4:].values
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

outlook[:,0] = le.fit_transform(outlook)
play[:,0] = le.fit_transform(play)

outlook
```

```python
ohe = preprocessing.OneHotEncoder()

outlook = ohe.fit_transform(outlook).toarray()
play = ohe.fit_transform(play).toarray()

df = pd.DataFrame(windy, columns=['windy'])
df_encoded = pd.get_dummies(df['windy'], prefix='windy')
windy_encoded = df_encoded.values
```

```python
windy_encoded
```

```python
sonuc1 = pd.DataFrame(outlook, columns=["overcast", "rainy", "sunny"])
sonuc2 = pd.DataFrame(play[:,0], columns=["play"])
sonuc3 = pd.DataFrame(windy_encoded[:,0], columns=["windy"])
```

```python
temperature = veriler[["temperature"]]
humidity = veriler[["humidity"]]
```

```python
s = pd.concat([sonuc1, sonuc2], axis=1)
s2 = pd.concat([s, sonuc3], axis=1)
s3 = pd.concat([s2, temperature], axis=1)
s4 = pd.concat([s3, humidity], axis=1)
```

```python
s4
```

# Model Oluşturma

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(s3, humidity, test_size=0.33, random_state=0)
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
y_pred
```
